# Topic: Password Strength Analyzer - Entropy calculation, regex complexity validations, and dictionary weak-word lookups
 
## 1. PASSWORD ENTROPY MATHEMATICS
- **Entropy**: Measures the unpredictability of a password. Represents the number of binary attempts (bits) a brute-force attacker requires to guess the password.
- **Formula (Hartley Entropy)**:
  $$E = L \log_2(R)$$
  - Where $L$ is the length of the password.
  - Where $R$ is the size of the character pool (charset range).
- **Common Character Pool Ranges ($R$)**:
  - Numeric digits (`0-9`): $R = 10$
  - Lowercase alphabets (`a-z`): $R = 26$
  - Uppercase alphabets (`A-Z`): $R = 26$
  - Special symbols (e.g. `!@#$...`): $R = 32$
  - Total combined characters pool: $R = 10 + 26 + 26 + 32 = 94$
- **Entropy Classifications**:
  - **< 40 bits**: Very Weak (easily brute-forced in seconds).
  - **40 - 59 bits**: Weak (subject to dictionary attacks).
  - **60 - 79 bits**: Medium (reasonable for non-sensitive apps).
  - **$\geq$ 80 bits**: Strong (highly secure).
 
## 2. COMPREHENSIVE ATTACK VECTORS
- **Dictionary Attack**: Attackers compare hashes against collections of known weak words (e.g., `"password"`, `"123456"`). A secure analyzer must cross-check passwords against a blacklist dictionary.
 
---


```python
import math
import re

class PasswordAnalyzer:
    # Small local dictionary of common weak passwords (for lab demonstration)
    COMMON_WEAK_PASSWORDS = {"password", "123456", "qwerty", "admin123", "letmein1"}

    def __init__(self, password):
        self.password = password
        self.length = len(password)

    def determine_pool_size(self):
        """Determines charset pool size R based on character classes present."""
        pool = 0
        if re.search(r"[0-9]", self.password):
            pool += 10
        if re.search(r"[a-z]", self.password):
            pool += 26
        if re.search(r"[A-Z]", self.password):
            pool += 26
        if re.search(r"[!@#$%^&*(),.?\":{}|<>]", self.password):
            pool += 32
        return pool

    def calculate_entropy(self):
        """Calculates password entropy: E = L * log2(R)."""
        pool_size = self.determine_pool_size()
        if pool_size == 0 or self.length == 0:
            return 0.0
        return self.length * math.log2(pool_size)

    def analyze(self):
        # 1. Check dictionary weak-list
        if self.password.lower() in self.COMMON_WEAK_PASSWORDS:
            return "Critical Warning: Common weak dictionary password!", 0.0
        
        entropy_val = self.calculate_entropy()
        
        # 2. Determine strength rating
        if entropy_val < 40:
            rating = "Very Weak (Low entropy)"
        elif entropy_val < 60:
            rating = "Weak"
        elif entropy_val < 80:
            rating = "Medium"
        else:
            rating = "Strong"
            
        return rating, entropy_val

print("--- Testing Password Strength Analyzer ---")
passwords_to_test = [
    "123456",            # Weak dictionary word
    "abcde",             # Low length, small pool
    "SafePassword123!",  # Large length, high pool variety
    "SecretPassword"     # Long but only lowercase/uppercase
]

for p in passwords_to_test:
    analyzer = PasswordAnalyzer(p)
    rating, entropy = analyzer.analyze()
    print(f"Password: {p:<18} | Rating: {rating:<15} | Entropy: {entropy:>6.2f} bits")
# Expected: "SafePassword123!" should show high entropy (>80 bits) and Strong rating
```
