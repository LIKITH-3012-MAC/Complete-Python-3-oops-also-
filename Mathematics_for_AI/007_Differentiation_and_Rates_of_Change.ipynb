# Topic: Differentiation and Rates of Change - Derivative rules, chain rule, and numerical difference quotients
 
## 1. DEFINITION OF THE DERIVATIVE
- **Derivative**: The instantaneous rate of change of a function at a given point, representing the slope of the tangent line to the function graph.
- **Mathematical Limit Definition**:
  $$f'(x) = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h}$$
- **Common Rules**:
  - Power Rule: $\frac{d}{dx} x^n = n x^{n-1}$
  - Product Rule: $\frac{d}{dx} [u(x)v(x)] = u'(x)v(x) + u(x)v'(x)$
  - Chain Rule: $\frac{d}{dx} f(g(x)) = f'(g(x)) \cdot g'(x)$
 
## 2. NUMERICAL DIFFERENTIATION
- When we do not have access to symbolic analytical equations, we approximate derivatives numerically using finite difference methods (where $h$ is a very small number, e.g. $10^{-5}$):
  1. **Forward Difference Quotient**:
     $$f'(x) \approx \frac{f(x + h) - f(x)}{h} \quad (\text{Error: } O(h))$$
  2. **Backward Difference Quotient**:
     $$f'(x) \approx \frac{f(x) - f(x - h)}{h} \quad (\text{Error: } O(h))$$
  3. **Central Difference Quotient**:
     $$f'(x) \approx \frac{f(x + h) - f(x - h)}{2h} \quad (\text{Error: } O(h^2))$$
     - *Note*: Central difference is numerically more accurate because its error scales quadratically ($O(h^2)$), cancelling out first-order error terms in the Taylor series expansion.
 
---


```python
import math

# Target function: f(x) = x^3 - 5x
# Analytical derivative: f'(x) = 3x^2 - 5
def f(x):
    return x**3 - 5*x

def f_prime_analytical(x):
    return 3 * (x**2) - 5

# 1. Finite Difference Implementations
def derivative_forward(func, x, h=1e-5):
    return (func(x + h) - func(x)) / h

def derivative_backward(func, x, h=1e-5):
    return (func(x) - func(x - h)) / h

def derivative_central(func, x, h=1e-5):
    return (func(x + h) - func(x - h)) / (2 * h)

print("--- Numerical Differentiation Benchmarks ---")
point = 2.0
true_val = f_prime_analytical(point)  # 3*(4) - 5 = 7.0

print(f"Analytical true derivative at x=2: {true_val}")

# Evaluate approximations
df_forward = derivative_forward(f, point)
df_backward = derivative_backward(f, point)
df_central = derivative_central(f, point)

print(f"Forward diff:  {df_forward:<10.8f} | Error: {abs(df_forward - true_val):.2e}")
print(f"Backward diff: {df_backward:<10.8f} | Error: {abs(df_backward - true_val):.2e}")
print(f"Central diff:  {df_central:<10.8f} | Error: {abs(df_central - true_val):.2e}")
# Expected: Central difference has significantly smaller error (around 10^-10 vs 10^-5)!

```

```python
# 2. Symbolic derivation using SymPy
import sympy as sp

print("\n--- Symbolic Chain Rule Derivation ---")
x_sym = sp.Symbol('x')
# Composite function: g(x) = sin(x^2)
g_sym = sp.sin(x_sym**2)

# Compute derivative
dg_sym = sp.diff(g_sym, x_sym)
print(f"g(x) expression: {g_sym}")
print(f"g'(x) derivative: {dg_sym}")  # Expected: 2*x*cos(x^2) (Chain rule)
```
