# Topic: Matrix Inverse and Systems - Gaussian elimination, condition numbers, and Moore-Penrose pseudo-inverse
 
## 1. SYSTEMS OF LINEAR EQUATIONS
- A system of linear equations is represented in matrix notation as:
  $$A\mathbf{x} = \mathbf{b}$$
  Where $A$ is the coefficient matrix, $\mathbf{b}$ is the output vector, and $\mathbf{x}$ is the unknown variable vector we solve for.
- **Gaussian Elimination**: An algorithm that performs elementary row operations (swapping, scaling, addition) to convert the augmented matrix $[A | \mathbf{b}]$ to upper triangular form, allowing back-substitution to solve for $\mathbf{x}$.
 
## 2. MATRIX INVERSE & CONDITION NUMBER
- **Matrix Inverse ($A^{-1}$)**:
  - Exists only for square matrices where $\det(A) \neq 0$ (non-singular).
  - Property: $AA^{-1} = A^{-1}A = I$.
  - Solution to system: $\mathbf{x} = A^{-1}\mathbf{b}$.
- **Condition Number ($\kappa(A)$)**:
  - Measures how sensitive a system is to numerical perturbations or rounding errors.
  - Formula: $\kappa(A) = \|A\| \cdot \|A^{-1}\|$.
  - **Well-conditioned**: $\kappa(A)$ close to 1. Solutions are stable.
  - **Ill-conditioned**: High $\kappa(A)$ value. Small changes in inputs trigger massive variations in output solutions.
 
## 3. MOORE-PENROSE PSEUDO-INVERSE (A+)
- In machine learning and data science, we often solve **overdetermined systems** (more equations than variables, $m > n$), where no exact solution exists.
- **Pseudo-inverse ($A^+$)**:
  - Computes the best least-squares fit solution minimizing $\|A\mathbf{x} - \mathbf{b}\|_2^2$.
  - For full column-rank matrices, it is calculated as:
    $$A^+ = (A^T A)^{-1} A^T$$
  - Least-squares solution: $\mathbf{x} = A^+\mathbf{b}$.
 
---


```python
import numpy as np

# 1. Solving Ax = b using standard numpy matrix inverse
print("--- Solving Ax = b ---")
# System:
#  x + 2y = 5
# 3x + 4y = 11
A = np.array([
    [1.0, 2.0],
    [3.0, 4.0]
])
b = np.array([5.0, 11.0])

# Compute inverse
A_inv = np.linalg.inv(A)
x = A_inv.dot(b)
print(f"Inverse A^-1:\n{A_inv}")
print(f"Solved x (using inverse): {x}")  # Expected: [1.0, 2.0] (1+4=5 | 3+8=11)

# Alternative standard solver (uses LU decomposition, numerically more stable)
x_solve = np.linalg.solve(A, b)
print(f"Solved x (using solve):   {x_solve}")

```

```python
# 2. Inspecting Condition Number
print("\n--- Testing Condition Number ---")
cond_A = np.linalg.cond(A)
print(f"Condition number of A: {cond_A:.4f}")  # Low value (~14.9), well-conditioned

# Create an ill-conditioned matrix
ill_matrix = np.array([
    [1.0, 1.0],
    [1.0, 1.00000001]
])
print(f"Condition number of ill-conditioned matrix: {np.linalg.cond(ill_matrix):.2e}")
# Expected: extremely high (~2.0e+08). Highly sensitive to rounding errors!

```

```python
# 3. Moore-Penrose Pseudo-inverse (Overdetermined System)
print("\n--- Moore-Penrose Pseudo-inverse (Least Squares) ---")
# Overdetermined system (3 equations, 2 unknowns):
#  x + y = 2
#  x - y = 1
# 2x + y = 3
A_over = np.array([
    [1.0, 1.0],
    [1.0, -1.0],
    [2.0, 1.0]
])
b_over = np.array([2.0, 1.0, 3.0])

# Compute pseudo-inverse directly: A^+ = (A^T A)^-1 A^T
A_pinv_manual = np.linalg.inv(A_over.T.dot(A_over)).dot(A_over.T)
x_pinv_manual = A_pinv_manual.dot(b_over)

# NumPy built-in pinv (uses SVD under the hood)
A_pinv_np = np.linalg.pinv(A_over)
x_pinv_np = A_pinv_np.dot(b_over)

print(f"Manual Pseudo-inverse A+:\n{A_pinv_manual}")
print(f"NumPy built-in pinv A+:\n{A_pinv_np}")
print(f"Least squares solution: {x_pinv_np}")
# Expected solution: ~[1.357, 0.571] which minimizes reconstruction error
```
