# Topic: Matrices and Linear Transformations - Coordinate mappings, multiplication axioms, rank, and determinants
 
## 1. MATRICES AS LINEAR TRANSFORMATIONS
- A **Matrix** is a rectangular grid array of numbers.
- Mathematically, an $m \times n$ matrix $A$ represents a **Linear Transformation** mapping vectors from $\mathbb{R}^n$ space to $\mathbb{R}^m$ space:
  $$T(\mathbf{x}) = A\mathbf{x}$$
- **Matrix-Vector Multiplication**: The result is a linear combination of the column vectors of $A$ weighted by the components of $\mathbf{x}$.
 
## 2. MATRIX MULTIPLICATION & PROPERTIES
- If $A$ is $m \times p$ and $B$ is $p \times n$, their product $C = AB$ is an $m \times n$ matrix, where:
  $$C_{ij} = \sum_{k=1}^p A_{ik} B_{kj}$$
- **Non-Commutative**: Matrix multiplication is **not commutative** ($AB \neq BA$ in general).
- **Identity**: The identity matrix $I$ acts as the multiplicative neutral element ($AI = IA = A$).
 
## 3. RANK, SPACES, AND DETERMINANTS
- **Rank**: The maximum number of linearly independent column (or row) vectors in the matrix.
- **Column Space (Range)**: The span (set of all linear combinations) of the matrix column vectors.
- **Null Space (Kernel)**: The set of all vectors $\mathbf{x}$ satisfying $A\mathbf{x} = \mathbf{0}$.
- **Determinant ($\det(A)$)**:
  - A scalar value mapping only for square matrices.
  - Geometrically, it represents the factor by which the linear transformation scales the area (2D) or volume (3D) of a unit region.
  - If $\det(A) = 0$, the transformation collapses space into a lower dimension, making the matrix singular (non-invertible).

```python
import numpy as np

# 1. Custom Matrix Multiplication from Scratch in pure Python
def multiply_matrices(matrix_a, matrix_b):
    """Multiplies two matrices from scratch.
    
    A is m x p, B is p x n, returns C of size m x n.
    """
    m = len(matrix_a)
    p1 = len(matrix_a[0])
    p2 = len(matrix_b)
    n = len(matrix_b[0])
    
    assert p1 == p2, "Inner dimensions must match for multiplication"
    
    # Initialize empty m x n matrix
    c = [[0.0] * n for _ in range(m)]
    
    # Compute dot products
    for i in range(m):
        for j in range(n):
            c[i][j] = sum(matrix_a[i][k] * matrix_b[k][j] for k in range(p1))
            
    return c

print("--- Custom Matrix Multiplication ---")
A = [
    [1, 2],
    [3, 4]
]
B = [
    [5, 6],
    [7, 8]
]

C_scratch = multiply_matrices(A, B)
print("A * B (Scratch):")
for row in C_scratch:
    print(row)
# Expected:
# [19, 22] (1*5 + 2*7 = 19 | 1*6 + 2*8 = 22)
# [43, 50] (3*5 + 4*7 = 43 | 3*6 + 4*8 = 50)

```

```python
# 2. Non-commutative validation (AB != BA)
print("\n--- Non-Commutativity Verification ---")
BA_scratch = multiply_matrices(B, A)
print("B * A (Scratch):")
for row in BA_scratch:
    print(row)
# Expected:
# [23, 34]
# [31, 46] (Proves AB != BA)

```

```python
# 3. Determinant and Rank using NumPy
print("\n--- Matrix Rank and Determinants (NumPy) ---")
A_np = np.array(A, dtype=float)

# Compute Determinant
det_A = np.linalg.det(A_np)
print(f"Determinant of A: {det_A:.4f}")  # Expected: -2.0 (1*4 - 2*3 = -2)

# Compute Rank
rank_A = np.linalg.matrix_rank(A_np)
print(f"Rank of A:        {rank_A}")      # Expected: 2 (Full rank)

# Singular matrix case (columns are linearly dependent)
singular_matrix = np.array([
    [1.0, 2.0],
    [2.0, 4.0]
])
print(f"\nSingular Matrix: \n{singular_matrix}")
print(f"Determinant:     {np.linalg.det(singular_matrix):.4f}")  # Expected: 0.0
print(f"Rank:            {np.linalg.matrix_rank(singular_matrix)}")  # Expected: 1
```
