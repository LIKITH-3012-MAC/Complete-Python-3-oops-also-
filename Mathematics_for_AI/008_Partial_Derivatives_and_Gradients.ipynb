# Topic: Partial Derivatives and Gradients - Directional derivatives, Jacobian/Hessian matrices, and steepest ascent properties
 
## 1. MULTIVARIATE CALCULUS: PARTIAL DERIVATIVES
- **Partial Derivative**: The derivative of a multi-variable function with respect to one variable, treating all other variables as constant:
  $$\frac{\partial f}{\partial x_i} = \lim_{h \to 0} \frac{f(x_1, \dots, x_i + h, \dots, x_n) - f(x_1, \dots, x_n)}{h}$$
 
## 2. THE GRADIENT VECTOR: steepest ascent direction
- **Gradient ($\nabla f(\mathbf{x})$)**: For a scalar-valued function $f: \mathbb{R}^n \to \mathbb{R}$, the gradient is the vector of all its partial derivatives:
  $$\nabla f(\mathbf{x}) = \left[ \frac{\partial f}{\partial x_1}, \frac{\partial f}{\partial x_2}, \dots, \frac{\partial f}{\partial x_n} \right]^T$$
- **Steepest Ascent**:
  - At any point $\mathbf{x}$, the gradient vector $\nabla f(\mathbf{x})$ points in the direction of the **steepest increase** (steepest ascent) of the function value.
  - The negative gradient $-\nabla f(\mathbf{x})$ points in the direction of the **steepest decrease** (steepest descent).
  - The magnitude $\|\nabla f(\mathbf{x})\|$ represents the rate of change in that direction.
- **Directional Derivative**: The rate of change of $f$ in the direction of a unit vector $\mathbf{v}$:
  $$D_{\mathbf{v}} f(\mathbf{x}) = \nabla f(\mathbf{x}) \cdot \mathbf{v}$$
 
## 3. JACOBIAN AND HESSIAN MATRICES
- **Jacobian Matrix ($J$)**: For a vector-valued function $\mathbf{f}: \mathbb{R}^n \to \mathbb{R}^m$, the Jacobian is the $m \times n$ matrix of first-order partial derivatives:
  $$J_{ij} = \frac{\partial f_i}{\partial x_j}$$
- **Hessian Matrix ($H$)**: For a scalar-valued function $f: \mathbb{R}^n \to \mathbb{R}$, the Hessian is the $n \times n$ symmetric matrix of second-order partial derivatives:
  $$H_{ij} = \frac{\partial^2 f}{\partial x_i \partial x_j}$$
  - Represents the local **curvature** of the function space. Used in Newton's optimization methods.

```python
import numpy as np

# 1. Defining a multi-variable function: f(x, y) = x^2 + 3xy + y^2
# Partial derivatives:
# df/dx = 2x + 3y
# df/dy = 3x + 2y
def f(coord):
    x, y = coord[0], coord[1]
    return x**2 + 3*x*y + y**2

def analytical_gradient(coord):
    x, y = coord[0], coord[1]
    return np.array([2*x + 3*y, 3*x + 2*y])

# 2. Numerical Gradient Estimation using Central Difference
def numerical_gradient(func, coord, h=1e-5):
    grad = np.zeros_like(coord)
    for i in range(len(coord)):
        coord_plus = coord.copy()
        coord_minus = coord.copy()
        
        coord_plus[i] += h
        coord_minus[i] -= h
        
        # Central difference along dimension i
        grad[i] = (func(coord_plus) - func(coord_minus)) / (2 * h)
    return grad

print("--- Numerical vs Analytical Gradients ---")
point = np.array([2.0, 3.0])  # Evaluate at (2, 3)

df_analytical = analytical_gradient(point)  # dx = 2*2 + 3*3 = 13 | dy = 3*2 + 2*3 = 12
df_numerical = numerical_gradient(f, point)

print(f"Analytical gradient at (2, 3): {df_analytical}")  # Expected: [13., 12.]
print(f"Numerical gradient at (2, 3):  {df_numerical}")   # Expected: [13., 12.]

```

```python
# 3. Directional Derivative Calculation
# Let's compute rate of change in direction of unit vector v = [0.8, 0.6] (3-4-5 triangle scaled)
v = np.array([0.8, 0.6])
# Ensure it is normalized
v = v / np.linalg.norm(v)

directional_derivative = np.dot(df_analytical, v)
print(f"\n--- Directional Derivative ---")
print(f"Direction Unit Vector: {v}")
print(f"Rate of change along v: {directional_derivative:.4f}")  # 13*0.8 + 12*0.6 = 17.6
```
