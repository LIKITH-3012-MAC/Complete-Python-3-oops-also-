# Topic: Vectors and Vector Spaces - Coordinate spans, dot product geometry, and vector norm metrics (L1, L2, L_inf)
 
## 1. VECTORS & SPACES
- **Vector Space**: A set of vectors closed under vector addition and scalar multiplication, satisfying 8 core axioms (associativity, commutativity, distributivity, etc.).
- **Linear Independence**: A set of vectors $\{\mathbf{v}_1, \dots, \mathbf{v}_n\}$ is linearly independent if the only scalar weights satisfying $\sum_{i=1}^n c_i \mathbf{v}_i = \mathbf{0}$ are $c_1 = \dots = c_n = 0$.
- **Basis**: A spanning set of linearly independent vectors for a vector space. The count of basis vectors determines the **Dimension** of the space.
 
## 2. DOT PRODUCT (INNER PRODUCT)
- **Algebraic Definition**: $\mathbf{u} \cdot \mathbf{v} = \sum_{i=1}^n u_i v_i$
- **Geometric Definition**: $\mathbf{u} \cdot \mathbf{v} = \|\mathbf{u}\| \|\mathbf{v}\| \cos(\theta)$
  - Where $\theta$ is the angle between the two vectors.
  - Orthogonal vectors have a dot product of `0` ($\cos(90^\circ) = 0$).
 
## 3. VECTOR NORMS (METRIC SPACES)
- A norm maps a vector to a non-negative real length value $\|\mathbf{x}\|$.
- **$L_1$ Norm (Manhattan Distance)**:
  $$\|\mathbf{x}\|_1 = \sum_{i=1}^n |x_i|$$
  - Sum of absolute values. Promotes sparsity in machine learning models (L1 regularization/Lasso).
- **$L_2$ Norm (Euclidean Distance)**:
  $$\|\mathbf{x}\|_2 = \sqrt{\sum_{i=1}^n x_i^2}$$
  - Standard geometric straight-line length.
- **$L_\infty$ Norm (Chebyshev / Maximum Norm)**:
  $$\|\mathbf{x}\|_\infty = \max_i |x_i|$$
  - Returns the absolute value of the largest single coordinate element.

```python
import math
import numpy as np

# 1. Custom Vector operations from scratch
class Vector:
    def __init__(self, coordinates):
        self.coords = tuple(float(x) for x in coordinates)
        self.dim = len(self.coords)

    def __repr__(self):
        return f"Vector({self.coords})"

    def add(self, other):
        """O(D) Vector Addition."""
        assert self.dim == other.dim, "Dimensions must match"
        return Vector(a + b for a, b in zip(self.coords, other.coords))

    def scalar_multiply(self, scalar):
        """O(D) Scalar Multiplication."""
        return Vector(scalar * x for x in self.coords)

    def dot(self, other):
        """O(D) Dot Product."""
        assert self.dim == other.dim, "Dimensions must match"
        return sum(a * b for a, b in zip(self.coords, other.coords))

    def norm_l1(self):
        """O(D) L1 Norm (Manhattan)."""
        return sum(abs(x) for x in self.coords)

    def norm_l2(self):
        """O(D) L2 Norm (Euclidean)."""
        return math.sqrt(sum(x**2 for x in self.coords))

    def norm_linf(self):
        """O(D) L-infinity Norm (Chebyshev)."""
        return max(abs(x) for x in self.coords)

print("--- Vector Operations from Scratch ---")
u = Vector([3, -4, 0])
v = Vector([1, 2, 2])

print(f"u: {u} | v: {v}")
print(f"u + v: {u.add(v)}")
print(f"u . v: {u.dot(v)}")  # 3*1 + -4*2 + 0*2 = -5

```

```python
# 2. Norm Metrics Verification
print("\n--- Vector Norm Metrics Verification ---")
print(f"u L1 Norm (Manhattan):   {u.norm_l1()}")   # |3| + |-4| + |0| = 7.0
print(f"u L2 Norm (Euclidean):   {u.norm_l2()}")   # sqrt(9 + 16) = 5.0
print(f"u L-inf Norm (Chebyshev): {u.norm_linf()}") # max(|3|, |-4|, |0|) = 4.0

```

```python
# 3. Validating with NumPy
print("\n--- NumPy Norm Validations ---")
u_np = np.array([3.0, -4.0, 0.0])

print(f"NumPy L1 norm:  {np.linalg.norm(u_np, ord=1)}")  # Expected: 7.0
print(f"NumPy L2 norm:  {np.linalg.norm(u_np, ord=2)}")  # Expected: 5.0
print(f"NumPy Linf norm: {np.linalg.norm(u_np, ord=np.inf)}")  # Expected: 4.0
```
