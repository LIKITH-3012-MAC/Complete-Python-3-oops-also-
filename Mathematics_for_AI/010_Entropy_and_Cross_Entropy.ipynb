# Topic: Entropy and Cross-Entropy - Information theory, Shannon entropy, KL divergence, and Softmax cross-entropy loss
 
## 1. INFORMATION THEORY FUNDAMENTALS
- **Self-Information**: Measures the "surprise" or information content of an event $x$.
  $$I(x) = -\log_2 P(x)$$
  - Highly probable events contain little information (low surprise); rare events contain high information.
 
## 2. SHANNON ENTROPY, CROSS-ENTROPY, AND KL DIVERGENCE
- **Shannon Entropy ($H(X)$)**: The expected value of information contained in a probability distribution $P$. Measures uncertainty.
  $$H(P) = -\sum_{i=1}^N P(x_i) \log_2 P(x_i)$$
- **Cross-Entropy ($H(P, Q)$)**: Measures the average number of bits needed to identify events from distribution $P$ using an encoding optimized for distribution $Q$ (the predicted model distribution).
  $$H(P, Q) = -\sum_{i=1}^N P(x_i) \log_2 Q(x_i)$$
  - Used as the default loss function for multi-class classification networks (Cross-Entropy Loss).
- **Kullback-Leibler (KL) Divergence ($D_{KL}(P \parallel Q)$)**:
  - Measures the extra bits required to encode $P$ using $Q$. Also known as the relative entropy.
  - Formula:
    $$D_{KL}(P \parallel Q) = H(P, Q) - H(P) = \sum_{i=1}^N P(x_i) \log_2 \frac{P(x_i)}{Q(x_i)}$$
  - **Properties**:
    - $D_{KL}(P \parallel Q) \geq 0$ (Gibbs' Inequality). Equal to 0 if and only if $P = Q$.
    - Non-symmetric: $D_{KL}(P \parallel Q) \neq D_{KL}(Q \parallel P)$ (not a true distance metric).
 
---


```python
import numpy as np

# 1. Calculating Shannon Entropy
def shannon_entropy(probs):
    """Calculates Shannon Entropy in bits (log base 2)."""
    probs = np.array(probs)
    # Mask out 0 probabilities to prevent log(0) errors
    nonzero_probs = probs[probs > 0]
    return -np.sum(nonzero_probs * np.log2(nonzero_probs))

print("--- Shannon Entropy ---")
fair_coin = [0.5, 0.5]
biased_coin = [0.9, 0.1]
certain_event = [1.0, 0.0]

print(f"Fair coin entropy:    {shannon_entropy(fair_coin):.4f} bits")      # Expected: 1.0000 (Maximum uncertainty)
print(f"Biased coin entropy:  {shannon_entropy(biased_coin):.4f} bits")    # Expected: ~0.4690 (Lower uncertainty)
print(f"Certain event entropy: {shannon_entropy(certain_event):.4f} bits") # Expected: 0.0000 (No uncertainty)

```

```python
# 2. Cross-Entropy Loss Calculation (from scratch)
def cross_entropy(p, q):
    """Calculates cross-entropy H(p, q) in bits (base 2)."""
    p = np.array(p)
    q = np.array(q)
    # Clip q to avoid division-by-zero or log(0) issues
    q = np.clip(q, 1e-15, 1.0)
    return -np.sum(p * np.log2(q))

print("\n--- Cross-Entropy Loss ---")
# True label: Class A [1, 0, 0]
true_label = [1.0, 0.0, 0.0]

# Model predictions
pred_good = [0.85, 0.10, 0.05]
pred_bad  = [0.20, 0.70, 0.10]

loss_good = cross_entropy(true_label, pred_good)
loss_bad  = cross_entropy(true_label, pred_bad)

print(f"Good model cross-entropy loss: {loss_good:.4f} bits")  # Lower loss
print(f"Bad model cross-entropy loss:  {loss_bad:.4f} bits")   # Higher loss

```

```python
# 3. KL Divergence Calculation
def kl_divergence(p, q):
    """Calculates KL Divergence D_KL(P || Q) in bits (base 2)."""
    p = np.array(p)
    q = np.array(q)
    # Clip values to prevent log(0) or division by zero
    p_clipped = np.clip(p, 1e-15, 1.0)
    q_clipped = np.clip(q, 1e-15, 1.0)
    
    return np.sum(p_clipped * np.log2(p_clipped / q_clipped))

print("\n--- KL Divergence relative distance ---")
print(f"D_KL(true || pred_good): {kl_divergence(true_label, pred_good):.4f} bits")
print(f"D_KL(true || pred_bad):  {kl_divergence(true_label, pred_bad):.4f} bits")
# Proves non-symmetric property: D_KL(P || Q) != D_KL(Q || P)
print(f"D_KL(pred_good || true): {kl_divergence(pred_good, true_label):.4f} bits")
```
