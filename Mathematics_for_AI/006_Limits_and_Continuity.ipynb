# Topic: Limits and Continuity - Formal definitions (epsilon-delta), limits computational rules, and symbolic evaluations
 
## 1. CORE PROPERTIES & MATH
- **Limit**: The value that a function approaches as the input approaches some value.
  $$\lim_{x \to c} f(x) = L$$
  - **Formal Definition ($\epsilon$-$\delta$)**: For every $\epsilon > 0$ there exists a $\delta > 0$ such that if $0 < |x - c| < \delta$, then $|f(x) - L| < \epsilon$.
  - **Right-hand / Left-hand Limit**: $\lim_{x \to c^+} f(x)$ and $\lim_{x \to c^-} f(x)$. The limit exists if and only if both side limits exist and are equal.
 
## 2. CONTINUITY & DISCONTINUITIES
- **Continuous Function**: A function $f(x)$ is continuous at $x = c$ if:
  1. $f(c)$ is defined.
  2. $\lim_{x \to c} f(x)$ exists.
  3. $\lim_{x \to c} f(x) = f(c)$.
- **Discontinuities**:
  - **Removable**: The limit exists, but $f(c)$ is either undefined or doesn't match the limit value (e.g. hole).
  - **Jump**: Left-hand and right-hand limits exist but are unequal (e.g. step function).
  - **Infinite**: The limit goes to $\pm \infty$ (e.g. vertical asymptote $f(x) = 1/x$ at $x=0$).

```python
import sympy as sp

# 1. Symbolic limit evaluation using SymPy
print("--- Symbolic Limits Evaluation ---")
x = sp.Symbol('x')

# Define a function with a removable discontinuity at x = 2
# f(x) = (x^2 - 4) / (x - 2)
f_expr = (x**2 - 4) / (x - 2)

# Evaluate limit as x approaches 2
limit_val = sp.limit(f_expr, x, 2)
print(f"Expression: (x^2 - 4) / (x - 2)")
print(f"Value at x=2 (naive): {f_expr.subs(x, 2)}")  # Expected: nan (0/0 division)
print(f"Limit as x -> 2:      {limit_val}")          # Expected: 4 (Removable discontinuity solved)

```

```python
# 2. Infinite Limit and Asymptote
print("\n--- Evaluating Infinite Limits ---")
g_expr = 1 / x

limit_right = sp.limit(g_expr, x, 0, dir='+')
limit_left  = sp.limit(g_expr, x, 0, dir='-')

print(f"Limit 1/x as x -> 0+ (right): {limit_right}")  # Expected: oo (infinity)
print(f"Limit 1/x as x -> 0- (left):  {limit_left}")   # Expected: -oo (-infinity)
# Since the left and right limits are different, the overall limit at x=0 does not exist!

```

```python
# 3. Limit of (sin(x) / x) as x -> 0
print("\n--- Fundamental Trigonometric Limit ---")
h_expr = sp.sin(x) / x
limit_h = sp.limit(h_expr, x, 0)
print(f"Limit sin(x)/x as x -> 0: {limit_h}")  # Expected: 1
```
