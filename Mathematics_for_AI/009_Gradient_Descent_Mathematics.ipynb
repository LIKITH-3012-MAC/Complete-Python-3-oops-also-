# Topic: Gradient Descent Mathematics - Convex optimization, updates iterations, and learning rate convergence proofs
 
## 1. OPTIMIZATION PARADIGM
- **Objective**: Find the parameter values $\mathbf{x}^*$ that minimize a cost function $f(\mathbf{x})$:
  $$\mathbf{x}^* = \arg\min_{\mathbf{x}} f(\mathbf{x})$$
 
## 2. CONVEXITY
- **Convex Function**: A function $f: \mathbb{R}^n \to \mathbb{R}$ is convex if for all $\mathbf{u}, \mathbf{v}$ and $\lambda \in [0, 1]$:
  $$f(\lambda \mathbf{u} + (1 - \lambda)\mathbf{v}) \leq \lambda f(\mathbf{u}) + (1 - \lambda)f(\mathbf{v})$$
  - Geometrically, this means any chord connecting two points on the function graph lies above or on the graph.
  - **Crucial Property**: For a convex function, **every local minimum is a global minimum**.
 
## 3. THE GRADIENT DESCENT ALGORITHM
- Since the negative gradient $-\nabla f(\mathbf{x})$ points in the direction of steepest descent, we can iteratively update our location starting from an initial point $\mathbf{x}^{(0)}$:
  $$\mathbf{x}^{(t+1)} = \mathbf{x}^{(t)} - \alpha \nabla f(\mathbf{x}^{(t)})$$
  Where $\alpha > 0$ is the **Learning Rate** (step size).
- **Learning Rate Impact**:
  - **Too Small**: Optimization requires many iterations to converge, risking getting stuck or wasting computation.
  - **Too Large**: Steps overshoot the minimum, causing oscillations, divergence, or numerical overflow.
 
---


```python
import numpy as np

# Let's minimize a simple convex 2D paraboloid: f(x, y) = x^2 + 4y^2
# Minimum is at (0, 0) where f(0, 0) = 0.
# Gradient: df/dx = 2x, df/dy = 8y
def f(coord):
    return coord[0]**2 + 4*coord[1]**2

def gradient_f(coord):
    return np.array([2 * coord[0], 8 * coord[1]])

# 1. Gradient Descent Optimization Loop
def optimize_gradient_descent(func_grad, start_point, learning_rate, max_iter=20):
    trajectory = []
    current_point = np.array(start_point, dtype=float)
    
    trajectory.append(current_point.copy())
    for i in range(max_iter):
        grad = func_grad(current_point)
        # Update equation: x_new = x_old - alpha * grad
        current_point -= learning_rate * grad
        trajectory.append(current_point.copy())
    return np.array(trajectory)

print("--- Gradient Descent Optimization ---")
start = [4.0, 3.0]

# Run with normal stable learning rate
trajectory_stable = optimize_gradient_descent(gradient_f, start, learning_rate=0.1, max_iter=15)

print("Trajectory coordinates:")
for idx, point in enumerate(trajectory_stable[:8]):
    print(f"Step {idx:2d} | Coordinates: [{point[0]:.4f}, {point[1]:.4f}] | Value: {f(point):.6f}")

# Verify convergence
final_point = trajectory_stable[-1]
print(f"Final coordinates after 15 steps: [{final_point[0]:.6f}, {final_point[1]:.6f}]")  # Close to (0, 0)

```

```python
# 2. Simulating Overshooting (Learning Rate too large)
print("\n--- Testing Learning Rate Overshooting ---")
# If learning rate is 0.3, the y coordinate update is: y = y - 0.3 * 8y = -1.4y.
# Since |-1.4| > 1, the y coordinate will oscillate and grow infinitely (divergence)!
trajectory_overshoot = optimize_gradient_descent(gradient_f, start, learning_rate=0.28, max_iter=5)

for idx, point in enumerate(trajectory_overshoot):
    print(f"Step {idx} | Coordinates: [{point[0]:.4f}, {point[1]:.4f}] | Value: {f(point):.2f}")
# Expected output shows oscillating coordinate growth!
```
