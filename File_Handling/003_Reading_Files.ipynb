# Topic: Reading Files - Stream extraction methods (read, readline, readlines) and lazy memory-efficient iterations
 
## 1. STREAM EXTRACTION METHODS
- **`f.read(size=-1)`**:
  - Reads and returns up to `size` characters (text mode) or bytes (binary mode).
  - If `size` is omitted or negative, it reads the entire remaining stream.
- **`f.readline(limit=-1)`**:
  - Reads and returns a single line from the file stream, including the newline character `\n`.
  - If `limit` is provided, reads at most `limit` characters.
- **`f.readlines(hint=-1)`**:
  - Reads all remaining lines in the file and returns them as a standard Python **list of strings**.
 
## 2. MEMORY HAZARD: read() & readlines() ON LARGE FILES
- **The Anti-Pattern**: Running `f.read()` or `f.readlines()` on a massive file (e.g. 10GB log file) attempts to load the entire content into RAM at once, triggering heap memory exhaustion and process crashes.
- **The Solution (Lazy Iteration)**:
  - The file object `f` is an **iterable** yielding lines lazily one at a time.
  - Writing `for line in f:` reads lines on demand using an internal buffer cache, keeping memory consumption at **$O(1)$ constant space** regardless of the file size.
 
---


```python
import os

# Create a demo file with multiple lines
demo_file = "reading_demo.txt"
with open(demo_file, "w", encoding="utf-8") as f:
    f.write("Line A: Hello World\nLine B: Python programming\nLine C: Performance matters\n")

# 1. Reading file in different styles
print("--- Using f.read() ---")
with open(demo_file, "r", encoding="utf-8") as f:
    # Read first 10 characters
    print(f"First 10 chars: {repr(f.read(10))}")
    # Read remaining content
    print(f"Remaining:      {repr(f.read())}")

```

```python
print("\n--- Using f.readline() ---")
with open(demo_file, "r", encoding="utf-8") as f:
    # Read line-by-line manually
    print(f"Line 1: {repr(f.readline().strip())}")
    print(f"Line 2: {repr(f.readline().strip())}")

```

```python
print("\n--- Using f.readlines() ---")
with open(demo_file, "r", encoding="utf-8") as f:
    # Returns a list of strings
    lines_list = f.readlines()
    print(f"Lines list: {lines_list}")

```

```python
# 2. Optimized Lazy Iteration for large files (Best Practice)
print("\n--- Lazy Line-by-Line Iteration (O(1) Space) ---")
with open(demo_file, "r", encoding="utf-8") as f:
    # Python uses a buffer behind the scenes; lines are yielded lazily on demand
    for line_number, line in enumerate(f, start=1):
        print(f"Processing Line {line_number}: {line.strip()}")

```

```python
# 3. Clean up
if os.path.exists(demo_file):
    os.remove(demo_file)
```
