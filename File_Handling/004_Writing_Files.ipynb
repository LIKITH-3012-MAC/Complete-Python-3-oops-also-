# Topic: Writing Files - Stream insertion methods (write, writelines) and encoding practices
 
## 1. STREAM INSERTION METHODS
- **`f.write(string)`**:
  - Writes the specified string (text mode) or bytes object (binary mode) to the stream.
  - Returns the count of characters or bytes written.
- **`f.writelines(iterable_collection)`**:
  - Writes an iterable collection of strings (or bytes) to the stream.
  - **Crucial Note**: `writelines` does **NOT** append newlines or separators automatically. It simply concatenates the items in the collection and writes them. You must add `\n` to your strings manually if you want lines.
 
## 2. BEST PRACTICE: ENCODING DEFINITION
- **The Risk**: If you do not specify an encoding when opening a text file, Python falls back to the system default encoding (`locale.getpreferredencoding()`).
  - This leads to platform-dependent bugs (e.g., a file written on Windows using CP-1252 might become corrupted when read on a Linux server using UTF-8).
- **The Solution**: Always explicitly specify the encoding parameter:
  `open("file.txt", "w", encoding="utf-8")`
 
---


```python
import os

write_demo_file = "writing_demo.txt"

# 1. Writing strings using write()
print("--- Using f.write() ---")
with open(write_demo_file, "w", encoding="utf-8") as f:
    chars_written_1 = f.write("Hello Python!\n")
    chars_written_2 = f.write("Writing file content.\n")
    print(f"Characters written in Call 1: {chars_written_1}")
    print(f"Characters written in Call 2: {chars_written_2}")

```

```python
# 2. Writing collections using writelines()
print("\n--- Using f.writelines() ---")
lines_to_write = [
    "Line 1: Item A\n",  # Notice explicit \n
    "Line 2: Item B\n",
    "Line 3: Item C"
]

# Overwrite mode 'w' replaces existing file
with open(write_demo_file, "w", encoding="utf-8") as f:
    f.writelines(lines_to_write)

# Verify the file contents
with open(write_demo_file, "r", encoding="utf-8") as f:
    print("File Content:")
    print(f.read())

```

```python
# 3. Showing the writelines separator omission trap
print("\n--- The writelines No-Newline Trap ---")
trap_lines = ["NoNewlines", "WillConcatenate", "InSingleLine"]

with open(write_demo_file, "w", encoding="utf-8") as f:
    f.writelines(trap_lines)

with open(write_demo_file, "r", encoding="utf-8") as f:
    print("Trap Result:")
    print(repr(f.read()))  # Expected: 'NoNewlinesWillConcatenateInSingleLine'

```

```python
# 4. Clean up
if os.path.exists(write_demo_file):
    os.remove(write_demo_file)
```
