# Topic: File Objects - Stream handle class hierarchies (io module), file descriptors, and stream validation
 
## 1. DEFINITION: FILE OBJECTS
- **File Object**: A Python object returned by the built-in `open()` function that acts as a **stream handle** (or file-like interface) to read from or write to a physical resource.
- **File Descriptor**: Under the hood, the OS kernel tracks open files using integer identifiers called File Descriptors (FDs). Python's file object wraps this FD integer (exposed via `f.fileno()`).
 
## 2. CPYTHON io MODULE HIERARCHY
- Depending on the mode passed to `open()`, CPython returns different classes from the standard `io` module library:
  1. **`io.TextIOWrapper`**: Returned for text modes (`r`, `w`, `a`, etc.). Manages encoding, decoding, and newline translation.
  2. **`io.BufferedReader`**: Returned for buffered binary read modes (`rb`).
  3. **`io.BufferedWriter`**: Returned for buffered binary write modes (`wb`, `ab`).
  4. **`io.FileIO`**: A low-level, unbuffered raw OS binary stream wrapped inside the buffered reader/writer classes.
 
## 3. STREAM VALIDATION METADATA
- File objects expose read-only attributes to validate state:
  - **`f.name`**: The path to the file.
  - **`f.mode`**: The access mode string.
  - **`f.closed`**: Boolean flag indicating if the stream has been closed.
 
---


```python
import io
import os

# 1. Inspecting file object class types
print("--- CPython File Object Class Hierarchy ---")

# Text Mode
with open("temp.txt", "w", encoding="utf-8") as f_text:
    print(f"Text mode class: {type(f_text).__module__}.{type(f_text).__name__}")
    # Expected: io.TextIOWrapper

# Binary Write Mode
with open("temp.bin", "wb") as f_bin_write:
    print(f"Binary write class: {type(f_bin_write).__module__}.{type(f_bin_write).__name__}")
    # Expected: io.BufferedWriter

# Binary Read Mode
with open("temp.bin", "rb") as f_bin_read:
    print(f"Binary read class:  {type(f_bin_read).__module__}.{type(f_bin_read).__name__}")
    # Expected: io.BufferedReader

```

```python
# 2. Inspecting File Descriptors and attributes
print("\n--- Stream Metadata and File Descriptors ---")
f = open("temp.txt", "r", encoding="utf-8")

print(f"File Name:       {f.name}")
print(f"Access Mode:     {f.mode}")
print(f"Is Closed?       {f.closed}")      # Expected: False
print(f"File Descriptor: {f.fileno()}")    # The OS integer identifier (e.g. 3 or 4)

f.close()
print(f"Is Closed after close()? {f.closed}")  # Expected: True

```

```python
# 3. Clean up temporary files
for file in ["temp.txt", "temp.bin"]:
    if os.path.exists(file):
        os.remove(file)
        print(f"Removed temp file: {file}")
```
