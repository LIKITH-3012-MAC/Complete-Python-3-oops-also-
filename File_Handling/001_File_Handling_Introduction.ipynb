# Topic: File Handling Introduction - Persistence concepts, text vs binary modes, and OS buffering
 
## 1. PERSISTENCE & DATA STORAGE
- **Volatile Memory (RAM)**: Fast execution storage that is cleared when the process exits or system shuts down.
- **Persistent Storage (Disk)**: Non-volatile storage (SSD, HDD) that retains data. Files are the standard abstraction used by operating systems to organize persistent disk data.
- **What is a File?**: At the physical layer, a file is simply a **sequence of raw bytes** stored on disk.
 
## 2. TEXT FILES VS BINARY FILES
1. **Text Files (`t` mode)**:
   - Interpreted as a sequence of characters.
   - Python automatically decodes bytes to strings using a character set encoding (usually UTF-8) when reading, and encodes strings back to bytes when writing.
   - Automatically normalizes newline delimiters (`\n` vs `\r\n`) depending on the OS platform (Universal Newline Support).
2. **Binary Files (`b` mode)**:
   - Accessed as raw, unmodified bytes (represented as Python `bytes` or `bytearray` objects).
   - No automatic character decoding or newline translation occurs. Used for images, audio, database files, and compiled code.
 
## 3. OPERATING SYSTEM BUFFERING
- Disk access is slow compared to RAM. To optimize performance, operating systems and Python use **Buffering**:
  - **Write Buffering**: Writes are accumulated in a memory cache. The OS writes the cached block to disk at once (flushing) rather than making slow disk writes for single characters.
  - **Read Buffering**: The OS reads a larger block of data from disk into a memory cache, serving subsequent read requests instantly from RAM.

```python
import os

# 1. Creating and writing a simple text file
print("--- Creating and Reading a Text File ---")
text_filename = "demo_text.txt"

with open(text_filename, "w", encoding="utf-8") as f:
    f.write("Line 1: Python is great!\nLine 2: File Handling intro.")

with open(text_filename, "r", encoding="utf-8") as f:
    content = f.read()
    print("Content read:")
    print(content)

```

```python
# 2. Creating and writing a binary file (raw bytes)
print("\n--- Creating and Reading a Binary File ---")
binary_filename = "demo_binary.bin"

# Write raw bytes
with open(binary_filename, "wb") as f:
    # Pass a bytes object (b"...")
    f.write(b"\x00\x01\x02\x03\x04\xff")

# Read raw bytes
with open(binary_filename, "rb") as f:
    byte_content = f.read()
    print(f"Bytes read: {byte_content} | Type: {type(byte_content).__name__}")
    # Print integer representation of bytes
    print(f"Raw integers: {list(byte_content)}")

```

```python
# 3. Clean up demo files
for file in [text_filename, binary_filename]:
    if os.path.exists(file):
        os.remove(file)
        print(f"Removed temporary file: {file}")
```
