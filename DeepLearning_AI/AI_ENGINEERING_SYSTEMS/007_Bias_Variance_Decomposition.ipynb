# 007: Bias-Variance Decomposition — Bias-variance tradeoff and noise derivation

## 1. THE DECOMPOSITION FORMULA
- The expected squared error of a model at a test point $x$ can be decomposed into:
  $$\mathbb{E}_D[(y - \hat{f}(x; D))^2] = \text{Bias}[\hat{f}(x)]^2 + \text{Variance}[\hat{f}(x)] + \sigma^2$$
  where:
  - **Bias** ($\text{Bias}[\hat{f}(x)] = \mathbb{E}_D[\hat{f}(x)] - f(x)$): Systematic error from wrong assumptions (underfitting).
  - **Variance** ($\text{Var}[\hat{f}(x)] = \mathbb{E}_D[(\hat{f}(x) - \mathbb{E}_D[\hat{f}(x)])^2]$): Sensitivity to fluctuations in the training dataset (overfitting).
  - **Noise** ($\sigma^2$): Irreducible error inherent in the data distribution.
---


```python
import numpy as np

class BiasVarianceSimulator:
    """Simulates bias and variance tradeoffs by fitting polynomials of different degrees."""
    @staticmethod
    def generate_data(n_samples=50, noise=0.2):
        X = np.sort(np.random.rand(n_samples) * 4 - 2)
        # Target: true function is a sine wave
        y = np.sin(X) + np.random.randn(n_samples) * noise
        return X, y

    @staticmethod
    def fit_and_evaluate(X_train, y_train, X_test, degree):
        # Fit polynomial
        coeffs = np.polyfit(X_train, y_train, degree)
        # Predict on test
        return np.polyval(coeffs, X_test)

```

```python
print("--- Bias-Variance Simulation ---")
np.random.seed(42)
X_test = np.linspace(-2, 2, 20)
y_test_true = np.sin(X_test)

# Run 100 trials with different dataset instances
n_trials = 100
degrees = [1, 3, 10]  # Underfit, Good fit, Overfit

for deg in degrees:
    predictions = []
    for _ in range(n_trials):
        X_tr, y_tr = BiasVarianceSimulator.generate_data(n_samples=30)
        pred = BiasVarianceSimulator.fit_and_evaluate(X_tr, y_tr, X_test, deg)
        predictions.append(pred)
        
    predictions = np.array(predictions)  # (n_trials, n_test_points)
    
    # Calculate bias^2 and variance
    mean_pred = np.mean(predictions, axis=0)
    bias_sq = np.mean((mean_pred - y_test_true) ** 2)
    variance = np.mean(np.var(predictions, axis=0))
    
    print(f"Degree {deg:2d} (Model complexity) → Bias^2: {bias_sq:.6f} | Variance: {variance:.6f}")

print("\n[Insight] Degree 1: High Bias, Low Variance (Underfit).")
print("[Insight] Degree 10: Low Bias, High Variance (Overfit).")
print("[Insight] Degree 3: Balanced tradeoff (optimal fit).")
```
