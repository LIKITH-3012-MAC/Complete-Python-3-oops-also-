# 008: Hyperparameter Optimization — Grid search, random search, and optimization loops

## 1. HYPERPARAMETER SEARCH
- **Grid Search**: Evaluates all combinations in a predefined grid. Suffers from the curse of dimensionality.
- **Random Search** (Bergstra & Bengio, 2012):
  - Samples hyperparameters randomly from a continuous probability distribution.
  - Empirically proven to find better configurations than Grid Search within fewer trials because most datasets have a low effective dimension (only a few hyperparameters significantly impact performance).

## 2. BAYESIAN OPTIMIZATION INTUITION
- Builds a surrogate probabilistic model of the objective function (e.g. Gaussian Process) to predict performance of configurations and selects the next point using an acquisition function to balance exploration and exploitation.
---


```python
import numpy as np

class RandomSearchOptimizer:
    """Implements random search hyperparameter optimization from scratch."""
    def __init__(self, parameter_space):
        """parameter_space: dict of parameter names to (min_val, max_val) ranges."""
        self.space = parameter_space
        self.results = []

    def sample_parameters(self):
        sampled = {}
        for param, (low, high) in self.space.items():
            sampled[param] = np.random.uniform(low, high)
        return sampled

    def search(self, evaluate_fn, n_trials=20):
        best_score = -np.inf
        best_params = None
        
        for trial in range(n_trials):
            params = self.sample_parameters()
            score = evaluate_fn(params)
            self.results.append((params, score))
            
            if score > best_score:
                best_score = score
                best_params = params
                
        return best_params, best_score

```

```python
# Define dummy evaluation function simulating validation accuracy
def mock_validation_accuracy(params):
    # Simulated optimal hyperparameters: learning_rate ≈ 0.01, regularization ≈ 0.005
    lr = params["learning_rate"]
    reg = params["regularization"]
    # Metric modeled as negative distance from optimal point + noise
    distance = (lr - 0.01) ** 2 + (reg - 0.005) ** 2
    accuracy = 0.95 - distance * 100 + np.random.randn() * 0.005
    return min(accuracy, 0.98)

# Run Random Search
print("--- Running Random Search Optimization ---")
np.random.seed(42)

param_ranges = {
    "learning_rate": (0.0001, 0.1),
    "regularization": (0.0001, 0.05)
}

optimizer = RandomSearchOptimizer(param_ranges)
best_config, best_acc = optimizer.search(mock_validation_accuracy, n_trials=15)

print("\nBest Configuration Found:")
print(f"  Learning Rate:  {best_config['learning_rate']:.5f}")
print(f"  Regularization: {best_config['regularization']:.5f}")
print(f"  Accuracy:       {best_acc:.4f}")
```
