# 006: ROC AUC Theory — ROC curve creation and probabilistic interpretation of AUC

## 1. ROC CURVE
- **Receiver Operating Characteristic (ROC)**: Plots True Positive Rate (TPR / Recall) vs False Positive Rate (FPR) at various probability thresholds:
  $$\text{TPR} = \frac{\text{TP}}{\text{TP} + \text{FN}}, \quad \text{FPR} = \frac{\text{FP}}{\text{FP} + \text{TN}}$$
- A random classifier produces a diagonal line from $(0,0)$ to $(1,1)$.

## 2. AREA UNDER THE CURVE (AUC)
- **Probabilistic Interpretation**: AUC is the probability that a classifier will rank a randomly chosen positive instance higher than a randomly chosen negative one.
  - $\text{AUC} = 1.0$: perfect separation.
  - $\text{AUC} = 0.5$: random guessing performance.
---


```python
import numpy as np

class ROCAUCEvaluator:
    """Computes ROC points and AUC from scratch using trapezoidal integration."""
    @staticmethod
    def compute_roc(y_true, y_probs):
        # Sort predictions and targets descending by probability score
        desc_indices = np.argsort(y_probs)[::-1]
        y_true_sorted = y_true[desc_indices]
        y_probs_sorted = y_probs[desc_indices]
        
        n_pos = np.sum(y_true)
        n_neg = len(y_true) - n_pos
        
        tpr_points = [0.0]
        fpr_points = [0.0]
        
        tp = 0
        fp = 0
        
        for i in range(len(y_true_sorted)):
            if y_true_sorted[i] == 1:
                tp += 1
            else:
                fp += 1
            tpr_points.append(tp / n_pos)
            fpr_points.append(fp / n_neg)
            
        return np.array(fpr_points), np.array(tpr_points)

    @classmethod
    def compute_auc(cls, y_true, y_probs):
        """Computes Area Under Curve (AUC) using the trapezoidal rule."""
        fpr, tpr = cls.compute_roc(y_true, y_probs)
        # Trapezoidal integration area under curve
        return np.trapz(tpr, fpr)

```

```python
# Testing the ROC-AUC evaluator
print("--- ROC-AUC Evaluation ---")
y_true = np.array([1, 0, 1, 0, 1, 0, 1, 0])
y_probs = np.array([0.9, 0.1, 0.8, 0.4, 0.7, 0.2, 0.85, 0.3])

fpr, tpr = ROCAUCEvaluator.compute_roc(y_true, y_probs)
auc = ROCAUCEvaluator.compute_auc(y_true, y_probs)

print(f"Computed AUC: {auc:.4f}")
print("FPR points: ", fpr.round(3))
print("TPR points: ", tpr.round(3))
```
