# 002: Perceptron and Convergence — The Perceptron learning rule and convergence theorem proof intuition

## 1. THEORY LAYER
- **Perceptron (Rosenblatt, 1958)**: The simplest trainable neural unit. A single-layer binary classifier.
- **Decision Rule**: Predict class $\hat{y} = 1$ if $\mathbf{w}^T \mathbf{x} + b \geq 0$, else $\hat{y} = 0$.
- **Learning Rule**: For each misclassified sample $(x_i, y_i)$:
  $$\mathbf{w} \leftarrow \mathbf{w} + \eta (y_i - \hat{y}_i) \mathbf{x}_i$$
  $$b \leftarrow b + \eta (y_i - \hat{y}_i)$$

## 2. CONVERGENCE THEOREM (Novikoff, 1962)
- **Statement**: If the training data is **linearly separable**, the Perceptron learning algorithm is guaranteed to converge (find a separating hyperplane) in a finite number of updates.
- **Proof Intuition**: The weight vector $\mathbf{w}$ can be decomposed into a component along the optimal separating direction $\mathbf{w}^*$ and an orthogonal residual. Each update increases alignment with $\mathbf{w}^*$ while the squared norm $\|\mathbf{w}\|^2$ grows at most linearly. The ratio $\frac{\mathbf{w} \cdot \mathbf{w}^*}{\|\mathbf{w}\|}$ increases monotonically, bounding the number of mistakes.
- **Limitation (Minsky & Papert, 1969)**: Perceptrons cannot learn non-linearly separable functions (e.g. XOR). This motivated multilayer networks.

---


```python
import numpy as np

class Perceptron:
    """Single-layer perceptron with the classic learning rule."""
    def __init__(self, n_features, learning_rate=1.0):
        self.weights = np.zeros(n_features)  # Initialize weights to zero
        self.bias = 0.0
        self.lr = learning_rate

    def predict(self, x):
        """Threshold activation: 1 if w·x + b >= 0, else 0."""
        z = np.dot(self.weights, x) + self.bias
        return 1 if z >= 0.0 else 0

    def train(self, X, y, max_epochs=100):
        """Train using the perceptron learning rule."""
        history = []
        for epoch in range(max_epochs):
            errors = 0
            for xi, yi in zip(X, y):
                y_hat = self.predict(xi)
                error = yi - y_hat
                if error != 0:
                    # Update rule: nudge weights toward correct classification
                    self.weights += self.lr * error * xi
                    self.bias += self.lr * error
                    errors += 1
            history.append(errors)
            if errors == 0:
                print(f"Converged at epoch {epoch + 1}")
                break
        return history

```

```python
# ── AND gate (linearly separable) ──
print("--- Training Perceptron on AND Gate ---")
X_and = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_and = np.array([0, 0, 0, 1])

p_and = Perceptron(n_features=2)
hist = p_and.train(X_and, y_and)

print("Predictions after training:")
for xi, yi in zip(X_and, y_and):
    print(f"  Input: {xi} | Target: {yi} | Predicted: {p_and.predict(xi)}")

```

```python
# ── XOR gate (NOT linearly separable) ──
print("\n--- Training Perceptron on XOR Gate ---")
X_xor = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_xor = np.array([0, 1, 1, 0])

p_xor = Perceptron(n_features=2)
hist_xor = p_xor.train(X_xor, y_xor, max_epochs=20)

print(f"Errors per epoch (last 5): {hist_xor[-5:]}")
print("Predictions (XOR cannot converge):")
for xi, yi in zip(X_xor, y_xor):
    print(f"  Input: {xi} | Target: {yi} | Predicted: {p_xor.predict(xi)}")
```
