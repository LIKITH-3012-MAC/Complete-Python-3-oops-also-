# Topic: Cyber Security Fundamentals - CIA Triad, AAA models, and quantitative risk calculators (ALE)
 
## 1. THE FOUNDATIONAL SECURITY MODELS
- **The CIA Triad**:
  1. **Confidentiality**: Ensuring data is accessible only to authorized users (mitigated via encryption, access controls).
  2. **Integrity**: Guaranteeing that data remains accurate and untampered (mitigated via hashing, digital signatures).
  3. **Availability**: Ensuring systems and resources are accessible on demand (mitigated via load balancing, backups, DDoS defenses).
- **The AAA Model**:
  - **Authentication**: Verifying *who* you are (e.g. passwords, biometrics, tokens).
  - **Authorization**: Determining *what* you can do (e.g. file system permissions, RBAC policies).
  - **Accounting (Auditing)**: Recording user actions and timestamps to establish accountability (e.g. secure system logs).
 
## 2. QUANTITATIVE RISK MATHEMATICS
- Security professionals calculate business risk using financial metrics:
  - **Asset Value (AV)**: Monetary value of the asset being protected.
  - **Exposure Factor (EF)**: Percentage of asset value lost in a single successful attack compromise.
  - **Single Loss Expectancy (SLE)**: Cost of a single security incident:
    $$\text{SLE} = \text{AV} \times \text{EF}$$
  - **Annualized Rate of Occurrence (ARO)**: The estimated number of times this specific threat succeeds per year.
  - **Annualized Loss Expectancy (ALE)**: The total estimated annual cost of this risk:
    $$\text{ALE} = \text{SLE} \times \text{ARO}$$
- This math helps companies budget security controls (a security system costing $50k/year is reasonable only if it reduces an ALE of $200k).
 
---


```python
class QuantitativeRiskCalculator:
    """A business security risk analyzer simulating enterprise risk assessments."""
    def __init__(self, asset_name, asset_value):
        self.asset_name = asset_name
        self.asset_value = float(asset_value)

    def calculate_risk(self, exposure_factor, annualized_rate_of_occurrence):
        """Computes SLE and ALE metrics."""
        # Exposure factor is a percentage representation (e.g., 0.25 for 25%)
        assert 0.0 <= exposure_factor <= 1.0, "Exposure factor must be between 0.0 and 1.0"
        
        sle = self.asset_value * exposure_factor
        ale = sle * annualized_rate_of_occurrence
        
        return {
            "asset": self.asset_name,
            "asset_value": self.asset_value,
            "exposure_factor": exposure_factor,
            "SLE": sle,
            "ARO": annualized_rate_of_occurrence,
            "ALE": ale
        }

```

```python
# Demonstrating Risk calculations for different scenarios
print("--- Quantitative Risk Assessment Analysis ---")

# Scenario A: E-commerce Server Database
# AV = $500,000. Successful SQLi leak database leads to a 40% loss (EF = 0.40).
# ARO = 0.5 (occurs once every 2 years).
db_calculator = QuantitativeRiskCalculator("Customer Database Server", 500000)
risk_a = db_calculator.calculate_risk(exposure_factor=0.40, annualized_rate_of_occurrence=0.5)

print(f"Asset:          {risk_a['asset']}")
print(f"Asset Value:    ${risk_a['asset_value']:,.2f}")
print(f"SLE (Single):   ${risk_a['SLE']:,.2f}")  # Expected: $200,000
print(f"ARO (Annual):   {risk_a['ARO']} events/year")
print(f"ALE (Expected): ${risk_a['ALE']:,.2f}")  # Expected: $100,000

```

```python
# Scenario B: Secure Budget Controls Cost Benefit
# Suppose a Web Application Firewall (WAF) costs $20,000/year to deploy.
# It reduces the ARO of database attacks from 0.5 to 0.05.
old_ale = risk_a['ALE']  # $100,000
new_risk = db_calculator.calculate_risk(exposure_factor=0.40, annualized_rate_of_occurrence=0.05)
new_ale = new_risk['ALE']  # $10,000

waf_cost = 20000
annual_savings = old_ale - new_ale
net_benefit = annual_savings - waf_cost

print("\n--- Cost-Benefit Analysis of Security Control (WAF) ---")
print(f"Previous ALE:   ${old_ale:,.2f}")
print(f"New ALE:        ${new_ale:,.2f}")
print(f"Annual Savings: ${annual_savings:,.2f}")
print(f"Control Cost:   ${waf_cost:,.2f}")
print(f"Net ROI Benefit: ${net_benefit:,.2f}")  # Expected: $70,000 savings
```
